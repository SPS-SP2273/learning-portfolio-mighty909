# 2 Better with NumPy

```python
# ok now instead of using lists we use arrays

import numpy as np

# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# convert to array
np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
```

```python
# numpy good because can plot trig functions straightaway

x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # The math stuff
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()          

# unfortunately this graph sucks because they dont have the same range in a sense
```

## 2.1 Adding mathematical functions

## We need another axis!

```python
# the commmand is called twinx()
# adds another y axis that shares the same x axis

import numpy as np

x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()     # This creates a new y-axis for the plots that come after
                                                            
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2

```

```python
# smoothening out the graph with np.linspace

x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()     # This creates a new y-axis for the plots that come after
                                                            
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')            
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24) # starts using data from row 25 & onwards
date = data[:, 0] # setting the x 
anomaly = data[:, -1] # and y

plt.plot(date, anomaly, alpha=.5) # for the plot 
plt.ylim([-8, 8]) # makes the plot symmetrical

# we can add a horizontal line at the zero value like so

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
#plt.hlines(y-value, starting x, ending x)
```

```python
# final product

plt.style.use('fivethirtyeight')

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
```
