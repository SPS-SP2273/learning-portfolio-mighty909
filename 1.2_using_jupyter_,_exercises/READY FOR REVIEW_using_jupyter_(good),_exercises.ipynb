```python
import timeit

squares1 = [i**2 for i in range(10)]

squares2 = list()
for i in range(10):
    squares2.append(i**2)

print(squares1, squares2)

print(timeit.timeit(stmt='[i**2 for i in range(10)]'))
print(timeit.timeit(stmt='squares2 = list();\nfor i in range(10): squares2.append(i**2)'))
```

```python
%%timeit

squares1 = [i**2 for i in range(10)]
```

```python
%%timeit

squares2 = list()
for i in range(10):
    squares2.append(i**2)
```

The numbers refer to the amount of time it takes to run the code in the cell. Specifically they are the mean and standard deviation of the sample of 7 runs of code.
